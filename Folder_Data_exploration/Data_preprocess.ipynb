```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

# Paths to your folders
folder_paths = [
    r'C:\Users\8897p\OneDrive\Desktop\NLP\Project\brainstorm\DATA\stories_cnn',
    r'C:\Users\8897p\OneDrive\Desktop\NLP\Project\brainstorm\DATA\stories_dail'
]

# Metadata storage
metadata = []

# Loop through folders and collect metadata
for folder_path in folder_paths:
    for root, _, files in os.walk(folder_path):
        for file in files:
            if file.endswith('.story'):
                file_path = os.path.join(root, file)
                file_size = os.path.getsize(file_path)
                modified_time = os.path.getmtime(file_path)
                word_count, char_count = 0, 0
                
                # Reading the file content to calculate word and character count
                with open(file_path, 'r', encoding='utf-8') as f:
                    content = f.read()
                    word_count = len(content.split())
                    char_count = len(content)
                
                metadata.append({
                    'folder': os.path.basename(folder_path),
                    'file_name': file,
                    'file_size_kb': file_size / 1024,  # size in KB
                    'modified_date': datetime.fromtimestamp(modified_time),
                    'word_count': word_count,
                    'char_count': char_count
                })

# Convert metadata to a DataFrame
df_metadata = pd.DataFrame(metadata)

# Save metadata to CSV to use later
df_metadata.to_csv('metadata_analysis.csv', index=False)
```

```python
# Sample display
print(df_metadata.head())
```

```python
# Data vizualization for CNN and Daily mail files count

import matplotlib.pyplot as plt

# Assign colors based on folder names 
colors = {
    'stories_cnn': '#1f77b4',      # Color for CNN
    'stories_dail': '#ff7f0e'      # Color for Daily Mail
}

# Create a color list based on the folder names in value_counts()
folder_counts = df_metadata['folder'].value_counts()
bar_colors = [colors[folder] for folder in folder_counts.index]

# Creating the bar chart
plt.figure(figsize=(9, 6))
ax = folder_counts.plot(
    kind='bar', 
    color=bar_colors, 
    title='File Count by Folder'
)

# Set labels for the x and y axes
plt.ylabel('Number of Files')
plt.xlabel('Folder')

# Add data labels on top of each bar
for i in ax.containers:
    ax.bar_label(i, label_type='edge', fontsize=10, color='black')

# Add the legend with colors for CNN and Daily Mail
handles = [
    plt.Line2D([0], [0], color='#1f77b4', lw=4, label='CNN'),
    plt.Line2D([0], [0], color='#ff7f0e', lw=4, label='Daily Mail')
]
plt.legend(handles=handles, loc='upper right', title='Folder Source')

# Show the plot
plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Seaborn grid
sns.set_style("whitegrid")

# Histogram for File Size Distribution
plt.figure(figsize=(10, 6))
sns.histplot(
    data=df_metadata,
    x='file_size_kb',
    bins=30,
    kde=True,  # Adding Kernel Density Estimate (smoothed curve)
    color='#1f77b4',
    edgecolor='black'
)
plt.title('File Size Distribution (KB)', fontsize=16)
plt.xlabel('File Size (KB)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

# Adding a legend and adjusting the layout
plt.legend(['File Size'], loc='upper right')
plt.tight_layout() 
plt.show()

# Histogram for Word Count Distribution with Seaborn
plt.figure(figsize=(10, 6))
sns.histplot(
    data=df_metadata,
    x='word_count',
    bins=30,
    kde=True,  # Adding Kernel Density Estimate (smoothed curve)
    color='#ff7f0e',
    edgecolor='black'
)
plt.title('Word Count Distribution', fontsize=16)
plt.xlabel('Word Count', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

# Adding a legend
plt.legend(['Word Count'], loc='upper right')
plt.tight_layout()  
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Grid setup
sns.set_style("whitegrid")

# Scatter Plot 1: Character Count vs. File Size
plt.figure(figsize=(10, 6))
sns.regplot(
    data=df_metadata,
    x='file_size_kb',
    y='char_count',
    scatter_kws={'alpha': 0.5, 's': 50},  
    line_kws={'color': 'red'},             # Regression line color
    color='blue'
)
plt.title('Character Count vs. File Size', fontsize=16) # title and lablel
plt.xlabel('File Size (KB)', fontsize=14)
plt.ylabel('Character Count', fontsize=14)
plt.tight_layout()
plt.show()

# Scatter Plot 2: Character Count vs. Word Count
plt.figure(figsize=(11, 5))
sns.scatterplot(
    data=df_metadata,
    x='word_count',
    y='char_count',
    hue='folder',            # Color-code by Cnn and Daily mail
    palette=['#1f77b4', '#ff7f0e'],  # Blue for CNN, Orange for Daily Mail
    alpha=0.3,               # scatter plot parameters
    s=50,                    
    edgecolor='black'        
)
sns.regplot(
    data=df_metadata,
    x='word_count',
    y='char_count',
    scatter=False,           
    line_kws={'color': 'green'},  # Regression line color
)
plt.title('Character Count vs. Word Count', fontsize=16)
plt.xlabel('Word Count', fontsize=14)
plt.ylabel('Character Count', fontsize=14)
plt.legend(title='Folder', loc='upper left')
plt.tight_layout()
plt.show()
```
